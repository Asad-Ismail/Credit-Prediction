```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report
```

```python
train_features=pd.read_csv("processed_data/train_features.csv")
train_labels=pd.read_csv("processed_data/train_labels.csv")
```

```python
val_features=pd.read_csv("processed_data/val_features.csv")
val_labels=pd.read_csv("processed_data/val_labels.csv")
```

```python
test_features=pd.read_csv("processed_data/test_features.csv")
test_labels=pd.read_csv("processed_data/test_labels.csv")
```

```python
# Flatten the labels DataFrames to 1D arrays
train_labels = train_labels.values.flatten()
val_labels = val_labels.values.flatten()
test_labels = test_labels.values.flatten()
```

```python
# Models
log_reg = LogisticRegression(solver='liblinear')
rf_clf = RandomForestClassifier()
xgb_clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
```

```python
log_reg_params = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}
rf_params = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
xgb_params = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}
```

```python
# Grid search with cross-validation
log_reg_gs = GridSearchCV(log_reg, log_reg_params, cv=3, scoring='roc_auc',verbose=3)
rf_gs = GridSearchCV(rf_clf, rf_params, cv=3, scoring='roc_auc',verbose=3)
xgb_gs = GridSearchCV(xgb_clf, xgb_params, cv=3, scoring='roc_auc',verbose=3)
```

```python
# Train Logistic Regression
log_reg_gs.fit(train_features, train_labels)
print("Best Logistic Regression Params: ", log_reg_gs.best_params_)
print("Best Logistic Regression Score: ", log_reg_gs.best_score_)
```

```python
# Train Random Forest
rf_gs.fit(train_features, train_labels)
print("Best Random Forest Params: ", rf_gs.best_params_)
print("Best Random Forest Score: ", rf_gs.best_score_)
```

```python
# Train XGBoost
xgb_gs.fit(train_features, train_labels)
print("Best XGBoost Params: ", xgb_gs.best_params_)
print("Best XGBoost Score: ", xgb_gs.best_score_)
```

```python
# Evaluate models on the validation set
log_reg_val_pred = log_reg_gs.predict(val_features)
rf_val_pred = rf_gs.predict(val_features)
xgb_val_pred = xgb_gs.predict(val_features)
```

```python
print("Logistic Regression Validation Accuracy: ", accuracy_score(val_labels, log_reg_val_pred))
print("Random Forest Validation Accuracy: ", accuracy_score(val_labels, rf_val_pred))
print("XGBoost Validation Accuracy: ", accuracy_score(val_labels, xgb_val_pred))
```

```python

# Evaluate the best model on the test set
# Assuming XGBoost performed best based on validation scores
best_model = xgb_gs.best_estimator_
test_pred = best_model.predict(test_features)
print("XGBoost Test Accuracy: ", accuracy_score(test_labels, test_pred))
print("XGBoost Test Classification Report: \n", classification_report(test_labels, test_pred))
```
